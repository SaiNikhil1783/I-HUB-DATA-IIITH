This lab will be based upon t-SNE which is a dimensionality reduction algorithm used to visualize high dimensional datasets.

t-SNE stands for t-Distributed Stochastic Neighbor Embedding. It is an unsupervised, non-linear technique primarily used for data exploration and visualizing high-dimensional data. It was developed by Laurens van der Maatens and Geoffrey Hinton in 2008 (Link to the paper: https://www.jmlr.org/papers/volume9/vandermaaten08a/vandermaaten08a.pdf)

image.png

t-SNE has a tuneable parameter, **perplexity** which balances attention between the local and global aspects of your data. It is a guess about the number of close neighbors each point has. The perplexity value has a complex effect on the resulting pictures. The original paper says, The performance of t-SNE is fairly robust to changes in the perplexity, and typical values are between 5 and 50. But the story is more nuanced than that. Getting the most from t-SNE may mean analyzing multiple plots with different perplexities.

Also the t-SNE algorithm doesnt always produce similar output on successive runs as there are additional hyperparameters related to the optimization process.

### HOW DOES T-SNE WORK??

The t-SNE algorithm calculates a similarity measure between pairs of instances in the high dimensional space and in the low dimensional space. It then tries to optimize these two similarity measures using a cost function. Lets break that down into 3 basic steps.

1. Measure similarities between points in the high dimensional space. Think of a bunch of data points scattered on a 2D space. For each data point (xi) well center a Gaussian distribution over that point. Then we measure the density of all points (xj) under that Gaussian distribution. Then renormalize for all points. This gives us a set of probabilities (Pij) for all points. Those probabilities are proportional to the similarities. All that means is, if data points x1 and x2 have equal values under this gaussian circle then their proportions and similarities are equal and hence you have local similarities in the structure of this high-dimensional space. The Gaussian distribution or circle can be manipulated using whats called perplexity, which influences the variance of the distribution (circle size) and essentially the number of nearest neighbors.

2. This step is similar to step 1, but instead of using a Gaussian distribution we use a Student t-distribution with one degree of freedom, which is also known as the Cauchy distribution (See fig below). This gives us a second set of probabilities (Qij) in the low dimensional space. As you can see the Student t-distribution has heavier tails than the normal distribution. The heavy tails allow for better modeling of far apart distances.

3. The last step is that we want these set of probabilities from the low-dimensional space (Qij) to reflect those of the high dimensional space (Pij) as best as possible. We want the two map structures to be similar. We measure the difference between the probability distributions of the two-dimensional spaces using Kullback-Liebler divergence (KL). Finally, we use gradient descent to minimize our KL cost function.

Scikit-learn has an implementation of t-SNE available which provides a wide variety of tuning parameters for t-SNE, and the most notable ones are:

1. n_components (default: 2): Dimension of the embedded space.
2. perplexity (default: 30): The perplexity is related to the number of nearest neighbors that are used in other manifold learning algorithms. Consider selecting a value between 5 and 50.
3. n_iter (default: 1000): Maximum number of iterations for the optimization. Should be at least 250.
4. method (default: barnes_hut): Barnes-Hut approximation runs in O(NlogN) time. method=exact will run on the slower, but exact, algorithm in O(N^2) time.

#### Visualizing the data in the projected space

---



```python
def scatter(x, colors):
    # We choose a color palette with seaborn.
    palette = np.array(sns.color_palette("hls", 10))

    # We create a scatter plot.
    f = plt.figure(figsize=(8, 8))
    ax = plt.subplot(aspect='equal')
    sc = ax.scatter(x[:,0], x[:,1], lw=0, s=40,
                    c=palette[colors.astype(int)])
    plt.xlim(-25, 25)
    plt.ylim(-25, 25)
    ax.axis('off')
    ax.axis('tight')

    # We add the labels for each digit.
    txts = []
    for i in range(10):
        # Position of each label.
        xtext, ytext = np.median(x[colors == i, :], axis=0)
        txt = ax.text(xtext, ytext, str(i), fontsize=24)
        txt.set_path_effects([
            PathEffects.Stroke(linewidth=5, foreground="w"),
            PathEffects.Normal()])
        txts.append(txt)
    plt.show()

    return f, ax, sc

scatter(digits_proj, y)
```

**1. Differences between PCA and t-SNE**

a. Which of the two algorithms is linear and which one is non-linear?

PCA (Principal Component Analysis):
 Linear algorithm. It finds directions (principal components) in the data that maximize variance and performs a linear transformation.

t-SNE (t-Distributed Stochastic Neighbor Embedding):
 Non-linear algorithm. It models high-dimensional relationships using probability distributions and maps them into a lower-dimensional space.

b. How does the non-linearity in one of these two algorithms help in capturing certain datasets?

t-SNE's non-linearity allows it to capture non-linear structures in the data, such as clusters, manifolds, or local relationships, which PCA cannot capture because it only identifies linear trends.

c. Does PCA preserve global distances, and does t-SNE do the same?

PCA: Preserves global structure. Points far apart in high-dimensional space remain far apart in lower-dimensional space.

t-SNE: Focuses on preserving local structure. It emphasizes maintaining neighborhood relationships while potentially distorting global distances.

d. Which one is computationally more expensive?

t-SNE: Computationally more expensive due to the pairwise distance calculations and optimization required to model the probability distributions.

PCA: Computationally efficient, especially with singular value decomposition (SVD), and scales better with dataset size.


```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.data
y = digits.target

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
tsne_result = tsne.fit_transform(X)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=y, cmap="Spectral", alpha=0.6)
plt.title("PCA: Global Structure Preserved")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label="Digit Class")

plt.subplot(1, 2, 2)
plt.scatter(tsne_result[:, 0], tsne_result[:, 1], c=y, cmap="Spectral", alpha=0.6)
plt.title("t-SNE: Local Neighborhoods Preserved")
plt.xlabel("t-SNE Component 1")
plt.ylabel("t-SNE Component 2")
plt.colorbar(label="Digit Class")

plt.tight_layout()
plt.show()
```

**2. Computational Complexity and Runtime of t-SNE**

Scalability:

t-SNE has a time complexity of O(N) for large datasets because it computes pairwise distances between all data points.

Using the Barnes-Hut approximation reduces complexity to O(N log N) for datasets with thousands of points, but it still becomes slow for very large datasets.

Factors Affecting Runtime:

Dataset size (N):
Larger datasets significantly increase computation time.

Dimensionality (D):
 High-dimensional data requires more computation to model pairwise probabilities.

Hyperparameters: Increasing the number of iterations or using higher perplexity values increases runtime.

Optimizations:

Pre-reduce dimensions using PCA or similar techniques before applying t-SNE.

Use GPU-accelerated libraries like FIt-SNE for large datasets.


```python
import numpy as np
import time
from sklearn.datasets import make_blobs

sizes = [500, 1000, 2000, 5000]
pca_times = []
tsne_times = []

for size in sizes:
    X, _ = make_blobs(n_samples=size, centers=5, n_features=20, random_state=42)

    start = time.time()
    pca = PCA(n_components=2)
    pca.fit_transform(X)
    pca_times.append(time.time() - start)

    start = time.time()
    tsne = TSNE(n_components=2, random_state=42, perplexity=30, n_iter=1000)
    tsne.fit_transform(X)
    tsne_times.append(time.time() - start)

plt.figure(figsize=(8, 6))
plt.plot(sizes, pca_times, label="PCA", marker='o')
plt.plot(sizes, tsne_times, label="t-SNE", marker='o')
plt.title("PCA vs. t-SNE Runtime")
plt.xlabel("Dataset Size")
plt.ylabel("Time (seconds)")
plt.legend()
plt.grid()
plt.show()
```

**3. Limitations or Potential Pitfalls of t-SNE**

**Loss of Global Structure:**

t-SNE prioritizes preserving local neighborhoods, often at the expense of global relationships.

Datasets with both local and global structures may not be well-represented.

**Sensitivity to Hyperparameters:**

Results are highly sensitive to parameters like perplexity and iterations.

Small changes can lead to vastly different visualizations, requiring trial-and-error tuning.

**Computational Expense:**


t-SNE is computationally intensive and scales poorly with large datasets, especially without dimensionality reduction.

**Non-deterministic Results:**

The algorithm relies on random initialization and may produce slightly different results across runs unless a fixed random seed is used.

**Interpretation Challenges: **

The axes of t-SNE plots are meaningless, and distances in low-dimensional space do not directly correspond to actual distances in high-dimensional space.
