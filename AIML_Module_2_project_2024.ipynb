**2. What are the 3 continuous features that show a significant statistical difference in distribution with respect to the presence and absence of the disease?**

The 3 continuous features with the most significant statistical difference are:

Oldpeak (p-value: 4.08e-15)

Thalach (p-value: 1.69e-14)

```python
import pandas as pd

file_path = 'heart.csv'
heart_data = pd.read_csv(file_path)

print(heart_data.head())
```

```python
from scipy.stats import ttest_ind

continuous_features = heart_data.drop(columns=['target', 'sex', 'cp', 'fbs', 'restecg', 'exang', 'ca', 'thal'])
significant_features = {}

for feature in continuous_features.columns:
    group_with_disease = heart_data[heart_data['target'] == 1][feature]
    group_without_disease = heart_data[heart_data['target'] == 0][feature]
    t_stat, p_value = ttest_ind(group_with_disease, group_without_disease)
    significant_features[feature] = p_value

sorted_significant_features = sorted(significant_features.items(), key=lambda x: x[1])
top_3_significant_features = sorted_significant_features[:3]
print(top_3_significant_features)
```

**3. Can we see a clear separation in terms of the presence/absence of the disease in the features obtained from PCA and t-SNE plots?**

Using PCA, there is some degree of separation, but it is not entirely distinct. The overlap indicates shared characteristics between samples with and without the disease.

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

pca = PCA(n_components=2)
pca_result = pca.fit_transform(continuous_features)
heart_data['pca_one'] = pca_result[:, 0]
heart_data['pca_two'] = pca_result[:, 1]

plt.figure(figsize=(8, 6))
for label, color in zip([0, 1], ['blue', 'red']):
    plt.scatter(
        heart_data[heart_data['target'] == label]['pca_one'],
        heart_data[heart_data['target'] == label]['pca_two'],
        label=f"Disease {label}",
        alpha=0.6
    )
plt.title("PCA: Disease Separation")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.grid()
plt.show()
```

**4. What is the optimal number of principal components in our case?**

Based on the explained variance ratio, the optimal number of components was calculated, but the cumulative variance for the first two components did not reach 90%. Thus, all components might be required to explain the variance effectively.

```python
explained_variance = pca.explained_variance_ratio_
cumulative_variance = explained_variance.cumsum()

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, marker='o')
plt.title("Cumulative Explained Variance by Principal Components")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.grid()
plt.show()
```

```python
##  one hot encoding of categorical features in data
def onehot_encode(df, columns, prefixes):
    df = df.copy()
    for column, prefix in zip(columns, prefixes):
        dummies = pd.get_dummies(df[column], prefix=prefix)
        df = pd.concat([df, dummies], axis=1)
        df = df.drop(column, axis=1)
    return df
```

```python
data = onehot_encode(
    data,
    columns=['Beverage', 'Beverage_prep'],
    prefixes=['bev', 'bevp']
)
```

```python
## to do
# perform pca on the data and plot the explained variace ratio, what is the optimal number of principal components in this case ?
```

The PCA plot indicates that 2–4 principal components are sufficient to explain over 90% of the variance. Reducing dimensions to 2–4 ensures most of the dataset’s variability is retained.



```python

numeric_columns = data.select_dtypes(include=['float64', 'int64']).drop(columns=['Tea'])  # Exclude 'Tea'
pca = PCA()
pca_result = pca.fit_transform(numeric_columns.fillna(0))  # Fill NaNs with 0
explained_variance_ratio = pca.explained_variance_ratio_

plt.figure(figsize=(10, 6))
plt.plot(range(1, len(explained_variance_ratio) + 1), explained_variance_ratio.cumsum(), marker='o')
plt.title("Cumulative Explained Variance Ratio")
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Variance Explained")
plt.grid()
plt.show()

optimal_components = next(i for i, total in enumerate(explained_variance_ratio.cumsum(), start=1) if total >= 0.9)
print(f"Optimal number of principal components: {optimal_components}")
```
